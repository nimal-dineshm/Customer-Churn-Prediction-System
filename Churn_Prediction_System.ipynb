```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
cd=pd.read_csv('/content/Telco_Customer_Churn_Dataset.csv')
cd1=pd.read_csv('/content/Telco_Customer_Churn_Dataset.csv')
cd2=cd1.copy()
cd3=cd1.copy()
```

```python
cd3.head()
```

```python
cd3.drop('customerID',axis=1,inplace=True)
```

```python
cd3['TotalCharges']=pd.to_numeric(cd3['TotalCharges'],errors='coerce')
```

```python
cd3['TotalCharges']=cd3['TotalCharges'].fillna(0)
```

```python
topic=['gender','Partner','Dependents','PhoneService','PaperlessBilling','Churn']
for col in topic:
  cd3[col]=LabelEncoder().fit_transform(cd3[col])
```

```python
cd3
```

```python
cd3=pd.get_dummies(cd3,drop_first=True)
```

```python
cd3
```

```python
bool_clmn = cd3.select_dtypes(include=['bool']).columns
```

```python
cd3[bool_clmn]=cd3[bool_clmn].astype(int)
```

```python
cd3
```

```python
cd3.shape
```

```python
x=cd3.drop('Churn',axis=1)
```

```python
y=cd3['Churn']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
sc=StandardScaler()
xtt=sc.fit_transform(X_train)
xtrest=sc.transform(X_test)
```

```python
model=RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
model.fit(xtt,y_train)
```

```python
pred=model.predict(xtrest)
```

```python
print(f"Accuracy Score: {accuracy_score(y_test, pred):.2%}")
print("\nConfusion Matrix:\n", confusion_matrix(y_test, pred))
print("\nDetailed Report:\n", classification_report(y_test, pred))
```
